**RESPOSTA**

Como pode ser visto, não há nenhum valor ausente - isso não significa que não há valores nulos, como 0. 
Contudo, há sim valores que são um pouco displicentes. Na tabela 2.1, percebe-se na coluna _Range_ que as seguintes variáveis apresentam: 
- `Pregnancies`: uma mulher teve 17 gravidezes. Apesar de não ser impossível, é incomum - a depender da época e região que o estudo foi feito 
- `Glucose`: segundo o [Ministério da Saúde](https://linhasdecuidado.saude.gov.br/portal/diabetes-mellitus-tipo-2-(DM2)-no-adulto/sou-paciente) e considerando que a unidade de medida é $mg/dl$, uma pessoa normal tem $100mg/dl$. Contudo, ter um valor 0 significa que essa pessoa está muito abaixo do normal ou, ainda por cima, não tem nenhuma glicose no corpo (mesmo que ainda seja a principal fonte de energia das células do corpo humano). Não é impossível, mas muito improvável.
- `BloodPressure`: segundo, novamente, o [Ministério da Saúde](https://www.gov.br/pt-br/noticias/saude-e-vigilancia-sanitaria/2021/05/saiba-como-descobrir-e-prevenir-a-hipertensao-arterial) e considerando a unidade de médidade de $hhMg$, uma pessoa saudável chega a 120mmHg. Logo, é um caso semelhante a variável anterior, onde existe uma pessoa com pressão arterial de 0.
- `Insulina`: sendo uma célula responsável pela redução da manutenção da glicose no sangue, faz sentido que haja medições com 0 (ainda mais para pessoas que desenvolveram diabetes)
- `BMI`: O IMC é uma fórmula representada por $\frac{peso}{altura²}$, existe registros iguais a 0. Logo, alguma pessoa há uma altura ou peso igual a 0. Dado impossível.
