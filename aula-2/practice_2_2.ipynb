# Atividade 2.2

Considere o conjunto de dados NHANES, disponibilizado pela biblioteca NHANES, o qual deriva do National Health and Nutrition Examination Survey, um estudo observacional de base populacional conduzido de forma contínua nos Estados Unidos. Esse inquérito tem como objetivo central avaliar, de maneira integrada, o estado de saúde, o perfil nutricional e as condições socioeconômicas da população civil não institucionalizada. Para atender a esse propósito, a base de dados reúne informações provenientes de entrevistas domiciliares, exames clínicos padronizados e análises laboratoriais, o que resulta em um conjunto amplo e intrinsecamente heterogêneo de variáveis. Essas variáveis abrangem aspectos demográficos, antropométricos, clínicos, laboratoriais e comportamentais, incluindo tanto atributos numéricos contínuos — como idade, peso corporal, estatura, índice de massa corporal e diferentes biomarcadores — quanto variáveis discretas e categóricas, a exemplo de sexo, raça/etnia, nível educacional e estado civil, além de apresentar valores ausentes associados ao próprio desenho amostral e à aplicação de distintos módulos de coleta. Diante dessas características, e utilizando o ambiente R, responda aos itens a seguir referentes à segunda etapa da AED, correspondente à sumarização dos dados.

1. Selecione um subconjunto de variáveis numéricas contínuas do banco de dados NHANES (por exemplo, idade, peso, altura, índice de massa corporal ou medidas laboratoriais). Para cada uma dessas variáveis:
- Obtenha medidas de tendência central.
- Calcule medidas de dispersão apropriadas.
- Identifique valores mínimos, máximos e intervalos.
- Indique quais funções do R são mais adequadas para essa sumarização inicial.

2. Para as variáveis selecionadas em (a), utilize funções de posição relativa para:
- Calcular quartis e percentis selecionados das variáveis numéricas.
- Obter o intervalo interquartílico (IQR).
- Explicar como essas medidas contribuem para a compreensão da variabilidade e da assimetria dos dados em comparação às medidas baseadas na média e no desvio-padrão.

3. Selecione pelo menos duas variáveis categóricas (por exemplo, sexo, raça/etnia, escolaridade ou estado civil). Para essas variáveis:
- Construa tabelas de frequências absolutas.
- Obtenha as correspondentes frequências relativas.

4. Para as variáveis selecionadas em (c), investigue a presença de valores ausentes, isto é,
- Identifique quais variáveis apresentam maior proporção de dados faltantes.
- Calcule o percentual de valores ausentes por variável.

5. Utilizando funções do pacote dplyr, realize uma sumarização estratificada, considerando:
- Ao menos três variável numérica.
- Um fator de estratificação (por exemplo, sexo ou grupo etário).

6. Construa tabelas de contingência para investigar a associação descritiva entre variáveis categóricas selecionadas no conjunto NHANES. Para isso:
- Selecione duas variáveis categóricas (por exemplo, Gender × Diabetes, ou Race1 × Education).
- Obtenha a tabela de frequências conjuntas por meio de table().
- Calcule proporções por linha e por coluna utilizando prop.table() com o argumento margin.
- Interprete, em termos descritivos, as diferenças nas distribuições condicionais entre categorias, evitando conclusões inferenciais formais.
- Discuta em que situações tabelas de contingência são mais informativas do que frequências marginais e quais limitações devem ser consideradas quando há categorias com baixa frequência.


7. Utilize a função mutate() (biblioteca dplyr) para criar variáveis e apoiar a sumarização exploratória. Considere, por exemplo:
- A criação de faixas etárias (por exemplo, AgeGroup) a partir de Age, definindo pontos de corte clinicamente interpretáveis.
- A criação de uma variável indicadora (binária) para sobrepeso/obesidade a partir de BMI, segundo um critério previamente definido.
- A recodificação de variáveis categóricas (por exemplo, agregando níveis raros ou reordenando fatores) com o objetivo de tornar as sumarizações mais interpretáveis.

8. Empregue a biblioteca summarytools para produzir sumarizações mais apresentáveis das variáveis selecionadas nos itens anteriores. Em particular:
- Utilize dfSummary() para obter uma visão integrada do conjunto (tipos de variáveis, estatísticas descritivas, frequências e proporção de valores ausentes).
- Aplique freq() para sumarizar variáveis categóricas, reportando frequências absolutas e relativas, bem como percentuais cumulativos quando pertinente.
- Utilize ctable() para construir tabelas cruzadas (contingência) com apresentação ampliada, incluindo percentuais (por linha, por coluna ou totais), conforme apropriado.
- Discuta as vantagens do uso do summarytools em relatórios reprodutíveis (por exemplo, clareza, padronização e controle de apresentação) em comparação às funções base do R (summary(), table(), prop.table()), especialmente em bases heterogêneas como NHANES.
