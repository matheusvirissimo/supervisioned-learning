### 1. Selecione um subconjunto de variáveis numéricas contínuas do banco de dados NHANES (por exemplo, idade, peso, altura, índice de massa corporal ou medidas laboratoriais). Para cada uma dessas variáveis:

```python
#O Subconjunto que iremos utilizar será do uso de drogas no mês
subset = ["DaysUsedMarijuanaOrHashishmonth", "DaysSmokedCigsDuringPast30Days", "DaysUsedMethamphetaminemonth", "OfDaysUsedCocainemonth"]
df_subset = data_df[subset]
```

```python
# medidas de tendência central - média, moda, mediana
print("===============\nMédia")
print(df_subset.mean())
print("===============\nMediana")
print(df_subset.median())
print("===============\nModa")
display(df_subset.mode())
```

```python
# Medidas de dispersão - variância/desvio padrão
print("=======\nVariância")
print(df_subset.var())
print("=======\nDesvio Padrão")
print(df_subset.std()) # standard
```

```python
# Mínimo, máximo e intervalos (considerando quartis)
intervals = [0.25, 0.50, 0.75]
print("===========\nMínimo")
print(df_subset.min())
print("===========\nMáximo")
print(df_subset.max())
```

### 2. Para as variáveis selecionadas em (a), utilize funções de posição relativa para:
- Calcular quartis e percentis selecionados das variáveis numéricas.
- Obter o intervalo interquartílico (IQR).
- Explicar como essas medidas contribuem para a compreensão da variabilidade e da assimetria dos dados em comparação às medidas baseadas na média e no desvio-padrão.
    
    **RESPOSTA**: Os quartis permitem ver a distribuição dos dados em proporções definidas. Por causa disso, podemos ter uma visão "local" sob os dados, o que permite pouca sensibilidade a _outliers_. Ao compararmos com outras métricas, como média, reparamos que há uma [continuar]

```python
quartile = [0.25, 0.50, 0.75]
print("===========\nIntervalos")
display(df_subset.quantile(quartile))
```

```python
q1 = df_subset.quantile(quartile[0])
q3 = df_subset.quantile(quartile[2])
iqr = q3 - q1

print(f"=============Interquartil\n {iqr}")
```
